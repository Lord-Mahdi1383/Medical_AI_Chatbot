```python
from optim.intel import OVModelForCasualLM
from transformers import AutoTokenizer
import os

# define directories
source_dir = "phi3_doctor_merged"
openvino_dir = "phi3_doctor_openvino_int8"
os.makedirs(openvino_dir, exist_ok=True)

# loading tokenizer
tokenizer = AutoTokenizer.from_pretrained(source_dir)

# converting and quantizing pytorch model to openvino int8
ov_model = OVModelForCasualLM.from_pretrained(
    source_dir,
    export=True,
    quantization_config={"weight_format": "int8"}
)

# saving 
ov_model.save_pretrained(openvino_dir)
tokenizer.save_pretrained(openvino_dir)
```
